# 키 / 몸무게 / 나이 numeric으로

```python
def splitter_h(x):
    if type(x) == str:
        arr = x.split('/')
        height = arr[0][:-2]
    else :
        height = 0
    return height
def splitter_w(x):
    if type(x) == str:
        arr = x.split('/')
        height = arr[1][:-2]
    else :
        height = 0
    return height

def split_birth(x):
    if type(x) == str:
        arr = x.split(" ")
        birthYear = arr[0][:-1]
    else :
        birthYear = 0
    return birthYear
```

```python
splitter_h('173kg/29kg')
```

```python
regular['height'] = regular['height/weight'].map(lambda x : splitter_h(x))
regular['weight'] = regular['height/weight'].map(lambda x : splitter_w(x))
regular['birth_year'] = regular['year_born'].map(lambda x : split_birth(x))
```
